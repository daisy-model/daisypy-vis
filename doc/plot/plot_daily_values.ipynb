Example illustrating how to make a scatter plot of daily logged variables.

See the scenarios in test-data/hourly

```python
import matplotlib.pyplot as plt
from daisy_vis.io import dlf
from daisy_vis.plot import plot_daily
```

```python
def daily_soil_chemical(hour, plot_line=False):
    '''Scatter plot of daily soil chemical for a single scenario'''
    dlfs = {
        'P2D' : dlf.read_dlf('../../test-data/hourly/P2D-Daily-Soil Chemical_110cm.dlf')
    }
    variables = [
        'Leak-Matrix',
        'Leak-Biopores',
        'Uptake',
        'Transform'
    ]
    plot_daily(dlfs, variables, hour=hour, title=f'Soil chemical @ hour {hour}', plot_line=plot_line)
```

Plot daily values measured at hour 0 with a line plot.

```python
daily_soil_chemical(0, True)
```

Plot daily values at hour 12 without a line plot

```python
daily_soil_chemical(12, False)
```
