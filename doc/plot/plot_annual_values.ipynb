Example illustrating how to make a bar plot of annualy logged variables for multiple scenarios.

See the scenarios in [../../test-data/annual](../../test-data/annual)

```python
import matplotlib.pyplot as plt
import daisy_vis as dv
```

```python
def annual_field_nitrogen():
    '''Bar plot of annual field nitrogen variables:
    Matrix-Leaching, Crop-Uptake, Soil-Drain and Surface-Loss
    for four scenarios

    Biopore-Leaching is not included as it is zero in all scenarios.
    '''
    dlfs = [
        dv.read_dlf(p) for p in [
            '../../test-data/annual/Annual-FN/HourlyP-Annual-FN-2.2b.dlf',
            '../../test-data/annual/Annual-FN/HourlyP-Annual-FN-2.3b.dlf',
            '../../test-data/annual/Annual-FN/HourlyP-Annual-FN-2.4b.dlf',
            '../../test-data/annual/Annual-FN/HourlyP-Annual-FN-2.5b.dlf'
        ]
    ]
    dlf_names = ['2.2b', '2.3b', '2.4b', '2.5b']
    variables = [
        'Matrix-Leaching',
        #'Biopore-Leaching', # There is no Biopore leaching in the scenarios
        'Crop-Uptake',
        'Soil-Drain',
        'Surface-Loss'
    ]
    dv.plot_annual(dlfs, variables, dlf_names=dlf_names, title='Annual Field Nitrogen')
    plt.show()
```

```python
def annual_tracer():
    '''Bar plot of annual tracer variables:
    Leak matrix, Uptake, Soil drain and Runoff
    for four scenarios

    Leak biopores is not included as it is zero in all scenarios.
    '''
    dlfs = [
        dv.read_dlf(p) for p in [
            '../../test-data/annual/Annual-Tracer/HourlyP-Annual-Tracer-2.2b.dlf',
            '../../test-data/annual/Annual-Tracer/HourlyP-Annual-Tracer-2.3b.dlf',
            '../../test-data/annual/Annual-Tracer/HourlyP-Annual-Tracer-2.4b.dlf',
            '../../test-data/annual/Annual-Tracer/HourlyP-Annual-Tracer-2.5b.dlf'
        ]
    ]
    variables = [
        'Leak matrix',
        # 'Leak biopores', # There is no Biopore leak in the scenarios
        'Uptake',
        'Soil drain',
        'Runoff'
    ]
    dlf_names = ['2.2b', '2.3b', '2.4b', '2.5b']
    dv.plot_annual(dlfs, variables, dlf_names=dlf_names, title='Annual Tracer')
    plt.show()
```

```python
annual_field_nitrogen()
```

```python
annual_tracer()
```
